```python
# NumPy is a Python library used for working with arrays.
# It also has functions for working in domain of linear algebra, fourier transform, and matrices.
# https://www.w3schools.com/python/numpy/numpy_intro.asp
import numpy as np

# Pandas is a Python library used for working with data sets.
# It has functions for analyzing, cleaning, exploring, and manipulating data.
# https://www.w3schools.com/python/pandas/pandas_intro.asp
import pandas as pd

# https://www.w3schools.com/python/matplotlib_intro.asp
import matplotlib.pyplot as plt
%matplotlib inline

# https://seaborn.pydata.org/tutorial/introduction.html
import seaborn as sns

dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning Experiment/Experiment 7/glass.csv")
```

```python
sns.pairplot(dataset, vars=["RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe"], hue="Type")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = np.array(dataset.iloc[:,1:])
Y = np.array(dataset["Type"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=23)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
for i in range(1, 25):
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(X_train, y_train)
  y_predict=knn.predict(X_test)
  acc=accuracy_score(y_test,y_predict)
  print(f'Accuracy Score of the Model: {acc} for {i}')
```

```python
knn = KNeighborsClassifier(n_neighbors = 8)
knn.fit(X_train,y_train)
```

```python
y_predict=knn.predict(X_test)
acc=accuracy_score(y_test,y_predict)
print(f'Accuracy Score of the Model: {acc}')
```

```python
print('Confusion Matrix :\n')
print(confusion_matrix(y_test,y_predict))
print(f'\nClassification Report \n\n {classification_report(y_test,y_predict)}')
```
